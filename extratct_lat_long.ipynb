```python
import csv
import requests

rest_api_key = 'bf0070cbed9ecd623aeead721c91397b'

# 주소를 입력 받아 위도와 경도를 반환하는 함수
def lat_long(address):
    # 카카오 로컬 API의 주소 검색 URL
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    # 카카오 API 인증 키를 헤더에 추가
    headers = {"Authorization": "KakaoAK " + rest_api_key}

    try:
        # API 요청을 보내고 응답을 받음
        response = requests.get(url, headers=headers)
        # JSON 형식의 응답 데이터를 파싱
        json_result = response.json()
        # JSON 데이터에서 주소 정보 추출
        address_xy = json_result['documents'][0]['address']
        # 경도(x)와 위도(y)를 실수형으로 반환
        return float(address_xy['x']), float(address_xy['y'])
    except Exception as e:
        # 예외 발생 시 에러 메시지를 출력하고 (None, None) 반환
        print(e)
        return None, None

# CSV 파일을 읽음
with open('./data/crisis(cp949).csv', 'r') as csvfile:
    # CSV 파일을 사전 형태로 읽음
    reader = csv.DictReader(csvfile)
    # 데이터를 리스트로 변환
    data = list(reader)

# 각 행에 위도와 경도 열 추가
for row in data:
    # 주소를 변수에 저장
    address = row['RONA_DADDR']
    # 주소를 이용해 위도와 경도 값 가져오기
    x, y = lat_long(address)
    # 행에 위도(x)와 경도(y) 값 추가
    row['x'] = x
    row['y'] = y

# CSV 파일의 필드명 리스트 생성
fieldnames = list(data[0].keys())
# 새로운 CSV 파일을 생성하고 데이터 저장
with open('./data/address.csv', 'w', newline='') as csvfile:
    # CSV 작성자를 필드명으로 초기화
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    # 헤더 작성
    writer.writeheader()
    # 데이터 작성
    writer.writerows(data)

# 완료 메시지 출력
print("CSV 파일이 위도와 경도 열과 함께 성공적으로 생성되었습니다!")
```
